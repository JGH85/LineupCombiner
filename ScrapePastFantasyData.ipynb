# Setup Function to import from url to dataframe

```python
def url_to_dataframe(url):
    print(url)
    response = requests.get(url)
    # Create BeautifulSoup object; parse with 'html.parser'
    soup = bs(response.text, 'html.parser')
    results = soup.find_all('pre')
    r = str(results[0])
    results_to_process = r[5:]
    results_to_process = results_to_process[:-6]
    csvstring2 = StringIO(results_to_process)
    df_to_return = pd.read_csv(csvstring2, sep=';')
#     print(df_to_return.head())
    return df_to_return
```

# Setup URLs

```python
# setup urls
base_url = 'http://rotoguru1.com/cgi-bin/fyday.pl?'
fd_search = 'scsv=1&game=fd'
dk_search = 'scsv=1&game=dk'
year_query = '&year='
week_query = '&week='
```

# Create list of 2018 urls for FD and DK

```python
url_list_fd_2018 = []
year = '2018'
for i in range (1,18):
    combined_url = base_url + fd_search + year_query + year + week_query + str(i)
    print(combined_url)
    url_list_fd_2018.append(combined_url)

url_list_dk_2018 = []
for i in range (1,18):
    combined_url = base_url + dk_search + year_query + year + week_query + str(i)
    print(combined_url)
    url_list_dk_2018.append(combined_url)
```

# Import from URLs to dataframes

```python
master_df_fd = pd.DataFrame()

for u in url_list_fd_2018:
    temp_df = url_to_dataframe(u)
    master_df_fd = master_df_fd.append(temp_df)
```

```python
master_df_dk = pd.DataFrame()

for u in url_list_dk_2018:
    temp_df = url_to_dataframe(u)
    master_df_dk = master_df_dk.append(temp_df)
```
