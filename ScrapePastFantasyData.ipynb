```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import csv
from io import StringIO


```

```python
url = 'http://rotoguru1.com/cgi-bin/fyday.pl?week=1&game=fd&scsv=1'
```

```python
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# print(soup.prettify())
```

```python
results = soup.find_all('pre')
# results
```

```python
r = str(results[0])
type(r)
```
